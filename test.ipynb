```python
import random
import math

home = "home"
enemy = "enemy"
event = "event"

class TreeConstructor:

    node_value = {
        home: 80,
        enemy : 30,
        event : 50
    }

    tendencies = {
        'pre_node': {
            'normal': 0.6,
            'branch': [0.3,0.5,0.4,0.2]
        },
        'node': {
            'home': 0.3,
            'enemy': 0.8,
            'event': 0.6
        }
    }

    branch_value = lambda self, count: (count-1)*30

    def __init__(self) -> None:
        self.tree_type = Tree
        self.target_len = 20
        self.max_width = 7
        self.min_width = 4
        self.curr_len = 1
        self.target_value = 600
        self.main_tree = self.tree_type(enemy)
        self.current_branches = [self.main_tree]

    def grow(self):
        next_branches = []
        for branch in self.current_branches:
            pre_node = random.choices(list(self.tendencies['pre_node'].keys()), weights=self.calc_branch_weight())[0]
            nodes = random.choices(list(self.tendencies['node'].keys()), weights=list(self.tendencies['node'].values()),
             k=1 if pre_node == 'normal' else 2)
            for node in nodes:
                new_node = Tree(node, self.branch_value(len(nodes))+self.node_value[node], branch)
                branch.child.append(new_node)
                next_branches.append(new_node)
        self.current_branches = next_branches[:]
        self.curr_len+= 1

    def calc_branch_weight(self):
        other = [self.tendencies['pre_node']['normal']]
        weights = self.tendencies['pre_node']['branch']
        if len(self.current_branches)>= self.max_width:
            return other+[0]
        curr_weight = weights[math.ceil(self.curr_len*(len(weights)+1)/self.target_len)-1]
        return other + [curr_weight * (self.min_width/len(self.current_branches))]

    def generate(self):
        if self.curr_len>= self.target_len:
            return self.main_tree
        self.grow()
        return self.generate()

class Tree:
    def __init__(self, node_type, extra_branching_value = 0, parent = None, child = None) -> None:
        self.type = node_type
        self.parent : Tree = parent
        self.child = child if child else []
        self.path_value = extra_branching_value
        if self.parent:
            self.path_value+= self.parent.path_value
            # self.parent.path_value + self.branch_value(len(self.parent.child))

    def __str__(self) -> str:
        return self.type

    def __repr__(self) -> str:
        return self.type+str(self.path_value)

def inspect(tree):
    toprint = [[tree]]
    x = 0
    while x<20:
        print(*toprint,sep='  ')
        temp = []
        for branch in toprint:
            for leaf in branch:
                temp.append(leaf.child)
        toprint = temp[:]
        x+= 1  
```

```python
testtree = TreeConstructor()
testtree.generate()

inspect(testtree.main_tree)
```
